# Quarterbacks EDA

```python
import pandas as pd
```

```python
qbs = pd.read_csv('../Capstone_Scrapping/Capstone_text_files/quarter_back.txt')
qbs.head()
```

```python
qbs['COMP%'] = qbs['COMP%'].str.replace('%', '').astype('float64')

qbs['YDS'] = qbs['YDS'].str.replace(',', '').astype('int64')
```

```python
qbs.rename(columns=lambda x: x.strip(), inplace = True)
```

```python
qbs.columns
```

```python
qbs.drop(columns = ['RTG', 'TEAM', 'GP'], inplace = True)
```

```python
qbs.head(1)
```

```python
qbs = qbs[qbs['ATT'] > 21].copy()
```

```python
qbs.dtypes
```

```python
qbs_stats = qbs.copy()
```

```python
qbs['COMP_rank'] = qbs['COMP'].rank(ascending = False)

qbs['ATT_rank'] = qbs['ATT'].rank(ascending = False)

qbs['ATT/G_rank'] = qbs['ATT/G'].rank(ascending = False)

qbs['YDS_rank'] = qbs['YDS'].rank(ascending = False)

qbs['300+ YDS_rank'] = qbs['300+ YDS'].rank(ascending = False)

qbs['COMP%_rank'] = qbs['COMP%'].rank(ascending = False)

qbs['YDS/ATT_rank'] = qbs['YDS/ATT'].rank(ascending = False)

qbs['YDS/COMP_rank'] = qbs['YDS/COMP'].rank(ascending = False)

qbs['TD_rank'] = qbs['TD'].rank(ascending = False)

qbs['INT_rank'] = qbs['INT'].rank()

qbs['QBR_rank'] = qbs['QBR'].rank(ascending = False)

qbs['LONG_rank'] = qbs['LONG'].rank(ascending = False)

qbs['SACKS_rank'] = qbs['SACKS'].rank()

qbs['SACK YDS_rank'] = qbs['SACK YDS'].rank()

qbs['FUML_rank'] = qbs['FUML'].rank()
```

```python
qbs.to_csv('../Capstone_csv_file/qbs_stats_rank_19-20', index = False)
```

```python
qbs_rank = qbs.drop(columns = ['COMP', 'ATT', 'ATT/G', 'YDS', '300+ YDS',
                               'COMP%', 'YDS/ATT', 'YDS/COMP', 'TD', 'INT',
                               'QBR', 'LONG', 'SACKS', 'SACK YDS', 'FUML'])
```

```python
qbs_rank['overall'] = (qbs['COMP_rank']+qbs['ATT_rank']+qbs['ATT/G_rank']+qbs['YDS_rank']+ 
                          qbs['300+ YDS_rank']+qbs['COMP%_rank']+qbs['YDS/ATT_rank']+
                          qbs['YDS/COMP_rank']+qbs['TD_rank']+qbs['QBR_rank']+qbs['LONG_rank']+
                          qbs['SACKS_rank']+qbs['SACK YDS_rank']+qbs['FUML_rank'])/14
```

```python
qbs_rank = qbs_rank.groupby('NAME', sort = False).mean()
```

```python
qbs_final_rank = {}
```

```python
qbs_final_rank['final_rank'] = qbs_rank['overall'].rank()
```

```python
qbs_final_rank = pd.DataFrame(qbs_final_rank)
```

```python
qbs_final_rank = qbs_final_rank.rank()
```

```python
qbs_final_rank = qbs_final_rank.sort_values('final_rank')
```
