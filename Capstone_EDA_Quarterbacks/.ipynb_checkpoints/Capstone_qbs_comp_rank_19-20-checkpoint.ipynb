# Quarterback Comp Rank

```python
import pandas as pd
```

```python
qbs = pd.read_csv('../Capstone_csv_file/qbs_stats_rank_19-20')
```

```python
qbs.head()
```

```python
qbs.rename(columns=lambda x: x.strip(), inplace = True)
```

```python
qbs = qbs[['NAME', 'COMP']].copy()
```

```python
qbs['COMP_rank'] = qbs['COMP'].rank(ascending = False)
```

```python
qbs = qbs.groupby('NAME', sort = False).mean()
```

```python
qbs = pd.DataFrame(qbs)
```

```python
qbs = qbs.sort_values('COMP_rank')
```

```python
qbs
```

```python
qbs.to_csv('../Capstone_csv_file/qbs_comp_rank_19-20')
```
