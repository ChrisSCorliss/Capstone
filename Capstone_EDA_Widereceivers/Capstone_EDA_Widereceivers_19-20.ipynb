# Widereceiver EDA

```python
import pandas as pd
```

```python
wrs = pd.read_csv('../Capstone_Scrapping/Capstone_text_files/wide_receiver.txt')
```

```python
wrs
```

```python
wrs.dtypes
```

```python
wrs['YDS'] = wrs['YDS'].str.replace(',', '').astype('int64')
```

```python
wrs.rename(columns=lambda x: x.strip(), inplace = True)
```

```python
wrs = wrs[wrs['TGT'] > 16].copy()
```

```python
wrs.columns
```

```python
wrs
```

```python
wrs = wrs.groupby('NAME', sort = False).mean()
```

```python
wrs.drop(columns = ['RTG', 'GP'], axis = 1, inplace = True)
```

```python
wrs_stats = wrs.copy()
```

```python
wrs_rank = {}
```

```python
wrs_rank['YDS_rank'] = wrs['YDS'].rank(ascending = False)
```

```python
wrs_rank['TGT_rank'] = wrs['TGT'].rank(ascending = False)
```

```python
wrs_rank['REC_rank'] = wrs['REC'].rank(ascending = False)
```

```python
wrs_rank['TGT/G_rank'] = wrs['TGT/G'].rank(ascending = False)
```

```python
wrs_rank['100+ YDS_rank'] = wrs['100+ YDS'].rank(ascending = False)
```

```python
wrs_rank['YDS/REC_rank'] = wrs['YDS/REC'].rank(ascending = False)
```

```python
wrs_rank['TD_rank'] = wrs['TD'].rank(ascending = False)
```

```python
wrs_rank['LONG_rank'] = wrs['LONG'].rank(ascending = False)
```

```python
wrs_rank['FUM_rank'] = wrs['FUM'].rank()
```

```python
wrs_rank['FUML_rank'] = wrs['FUML'].rank()
```

```python
wrs_rank['overall'] = (wrs_rank['TGT_rank'] + wrs_rank['REC_rank'] + wrs_rank['TGT/G_rank'] + wrs_rank['100+ YDS_rank'] +
            wrs_rank['YDS/REC_rank'] + wrs_rank['TD_rank'] + wrs_rank['LONG_rank'] +
            wrs_rank['FUM_rank'] + wrs_rank['FUML_rank'] + wrs_rank['YDS_rank'])/10
```

```python
wrs_rank = pd.DataFrame(wrs_rank)
```

```python
wrs_final_rank = {}
```

```python
wrs_final_rank['final_rank'] = wrs_rank['overall'].rank()
```

```python
wrs_final_rank = pd.DataFrame(wrs_final_rank)
```

```python
wrs_final_rank = wrs_final_rank.sort_values('final_rank')
```

```python
wrs_final_rank.to_csv('../Capstone_csv_file/wrs_final_rank_19-20')
```

```python
wrs = wrs_stats.merge(wrs_rank, on = 'NAME')
```

```python
wrs.to_csv('../Capstone_csv_file/wrs_stats_rank_19-20')
```
