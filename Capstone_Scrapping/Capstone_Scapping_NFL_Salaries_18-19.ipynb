```python
soup_salaries = BeautifulSoup(driver.page_source)
```

```python
soup_salaries
```

```python
table = soup_salaries.find('tbody', attrs = {'aria-relevant': 'all'})
salaries_18 = []
for row in table.find_all('tr', attrs = {'role': 'row'}):
    time.sleep(randrange(3,15))
    salary = {}
    salary['name'] = row.find('a', attrs = {'class': 'team-name'}).text
    salary['salary'] = row.find('span', attrs = {'class': 'info'}).text
    salaries_18.append(salary)
pd.DataFrame(salaries_18)
```

```python
salaries = pd.DataFrame(salaries_18)
```
