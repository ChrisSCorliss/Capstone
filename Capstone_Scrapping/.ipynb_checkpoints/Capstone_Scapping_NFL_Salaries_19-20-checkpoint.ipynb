# Scrapped Salaries from Spotrac

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from random import randrange
```

```python
driver = webdriver.Chrome(executable_path="./chromedriver capstone/chromedriver")
```

```python
url = 'https://www.spotrac.com/nfl/rankings/average//'
driver.get(url)
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
soup_salaries = BeautifulSoup(driver.page_source)
```

```python
soup_salaries
```

```python
table = soup_salaries.find('tbody', attrs = {'aria-relevant': 'all'})
salaries = []
for row in table.find_all('tr', attrs = {'role': 'row'}):
    time.sleep(randrange(3,15))
    salary = {}
    salary['name'] = row.find('a', attrs = {'class': 'team-name'}).text
    salary['salary'] = row.find('span', attrs = {'class': 'info'}).text
    salaries.append(salary)
pd.DataFrame(salaries)
```

```python
salaries = pd.DataFrame(salaries)
```

```python
salaries.head()
```

```python
salaries.to_csv('salaries.csv', index = False)
```
