# Scrapping Lineups for Tight Ends Statistics

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from random import randrange
```

```python
driver = webdriver.Chrome(executable_path="./chromedriver capstone/chromedriver")
```

```python
url = 'https://www.lineups.com/nfl/player-stats/tight-end-te-stats'
driver.get(url)
```

```python
res = requests.get(url)
```

```python
res.status_code
```

### Before you start the next line of code you must go to the driver page and click the button to switch it from table to csv mode

```python
soup_tight_end = BeautifulSoup(driver.page_source)
```

```python
soup_tight_end
```

```python
with open('./Capstone_text_files/tight_end.txt', 'w') as te_out:
    for i in soup_tight_end.find_all('p', attrs = {'csv-text'}):
        tes = i.text
        te_out.write(tes)
    
 
```
