# Scrapping Lineups for Wide Receiver Statistics

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from random import randrange
```

```python
driver = webdriver.Chrome(executable_path="./chromedriver capstone/chromedriver")
```

```python
url = 'https://www.lineups.com/nfl/player-stats/wide-receiver-wr-stats'
driver.get(url)
```

```python
res = requests.get(url)
```

```python
res.status_code
```

### Before you start the next line of code you must go to the driver page and click the button to switch it from table to csv mode

```python
soup_receiving = BeautifulSoup(driver.page_source)
```

```python
soup_receiving
```

```python
with open('./Capstone_text_files/wide_receiver.txt', 'w') as wr_out:
    for i in soup_receiving.find_all('p', attrs = {'csv-text'}):
        wrs = i.text
        wr_out.write(wrs)
    
 
```
