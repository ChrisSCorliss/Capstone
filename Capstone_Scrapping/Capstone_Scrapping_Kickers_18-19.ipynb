# Scrapping Lineups for Field Goal Kickers Statistics

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from random import randrange
```

```python
driver = webdriver.Chrome(executable_path="./chromedriver capstone/chromedriver")
```

```python
url = 'https://www.lineups.com/nfl/player-stats/kicker-stats'
driver.get(url)
```

```python
res = requests.get(url)
```

```python
res.status_code
```

### Before you start the next line of code you must go to the driver page and click the button to switch it from table to csv mode

```python
soup_kicker = BeautifulSoup(driver.page_source)
```

```python
soup_kicker
```

```python
with open('./Capstone_text_files/kicker_18-19.txt', 'w') as kick_out:
    for i in soup_kicker.find_all('p', attrs = {'csv-text'}):
        kicks = i.text
        kick_out.write(kicks)
    
```
