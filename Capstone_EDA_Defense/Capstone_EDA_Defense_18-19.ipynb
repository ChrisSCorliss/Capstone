# EDA of Defensive Statistics

```python
import pandas as pd
```

```python
defs = pd.read_csv('../Capstone_Scrapping/Capstone_text_files/defense_18-19.txt')
```

```python
defs.head()
```

```python
defs.rename(columns=lambda x: x.strip(), inplace = True)
```

```python
defs_1 = defs.drop(['RTG', 'TEAM', 'GP'], axis = 1)
```

```python
defs_1.head()
```

```python
defs_1['TCKL_rank'] = defs_1['TCKL'].rank(ascending = False)
```

```python
defs_1['SOLO_rank'] = defs_1['SOLO'].rank(ascending = False)
```

```python
defs_1['AST_rank'] = defs_1['AST'].rank(ascending = False)
```

```python
defs_1['TFL_rank'] = defs_1['TFL'].rank(ascending = False)
```

```python
defs_1['SCK_rank'] = defs_1['SCK'].rank(ascending = False)
```

```python
defs_1['SCK YDS_rank'] = defs_1['SCK YDS'].rank(ascending = False)
```

```python
defs_1['QBH_rank'] = defs_1['QBH'].rank(ascending = False)
```

```python
defs_1['PASS DEF_rank'] = defs_1['PASS DEF'].rank(ascending = False)
```

```python
defs_1['FF_rank'] = defs_1['FF'].rank(ascending = False)
```

```python
defs_1['FUMR_rank'] = defs_1['FUMR'].rank(ascending = False)
```

```python
defs_1['FUMR YDS_rank'] = defs_1['FUMR YDS'].rank(ascending = False)
```

```python
defs_1['FUMR TD_rank'] = defs_1['FUMR TD'].rank(ascending = False)
```

```python
defs_1['INT_rank'] = defs_1['INT'].rank(ascending = False)
```

```python
defs_1['INT YDS_rank'] = defs_1['INT YDS'].rank(ascending = False)
```

```python
defs_1['INT TD_rank'] = defs_1['INT TD'].rank(ascending = False)
```

```python
defs_1
```

```python
defs_1_rank = defs_1.drop(['TCKL', 'SOLO', 'AST', 'TFL', 'SCK', 'SCK YDS', 
             'QBH', 'PASS DEF', 'FF', 'FUMR', 'FUMR YDS', 'FUMR TD',
             'INT', 'INT YDS', 'INT TD'], axis = 1)           
```

```python
defs_1_rank.head()
```

```python
defs_1_rank.columns
```

```python
defs_1_rank_avg = defs_1_rank.groupby('NAME', sort = False).mean()
```

```python
defs_1_rank_avg
```

```python
defs_1_rank_avg['overall'] = (defs_1_rank_avg['TCKL_rank'] + defs_1_rank_avg['SOLO_rank'] + defs_1_rank_avg['AST_rank'] + defs_1_rank_avg['TFL_rank'] +
defs_1_rank_avg['SCK_rank'] + defs_1_rank_avg['SCK YDS_rank'] + defs_1_rank_avg['QBH_rank'] + defs_1_rank_avg['PASS DEF_rank'] +
defs_1_rank_avg['FF_rank'] + defs_1_rank_avg['FUMR_rank'] + defs_1_rank_avg['FUMR YDS_rank'] + defs_1_rank_avg['FUMR TD_rank'] +
defs_1_rank_avg['INT_rank'] + defs_1_rank_avg['INT YDS_rank'] + defs_1_rank_avg['INT TD_rank']) / 15
```

```python
defs_1_rank_avg
```

```python
defs_overall_rank = {}
```

```python
defs_overall_rank['overall'] = defs_1_rank_avg['overall']
```

```python
defs_overall_rank = pd.DataFrame(defs_overall_rank)
```

```python
defs_overall_rank['final_rank'] = defs_overall_rank['overall'].rank()
```

```python
defs_final_overall_rank = defs_overall_rank['final_rank']
```

```python
defs_final_overall_rank = pd.DataFrame(defs_final_overall_rank)
```

```python
defs_final_overall_rank = defs_final_overall_rank.sort_values('final_rank')
```

```python
defs_final_overall_rank
```

```python
defs_final_overall_rank.to_csv('../Capstone_csv_file/defense_final_rank_18-19.csv', index = False)
```
