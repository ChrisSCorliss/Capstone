```python
import pandas as pd
```

```python
rbs = pd.read_csv('../Capstone_Scrapping/Capstone_text_files/running_backs.txt')
```

```python
rbs
```

```python
rbs['YDS'] = rbs['YDS'].str.replace(',', '').astype('int64')
```

```python
rbs.rename(columns=lambda x: x.strip(), inplace = True)
```

```python
rbs = rbs[rbs['ATT'] > 16].copy()
```

```python
rbs
```

```python
rbs.drop(columns = ['RTG', 'TEAM', 'GP'], inplace = True)
```

```python
rbs = rbs.groupby('NAME', sort= False).mean()
```

```python
rbs.dtypes
```

```python
rbs_stats = rbs.copy()
```

```python
rbs['YDS_rank'] = rbs['YDS'].rank(ascending = False)
```

```python
rbs['ATT_rank'] = rbs['ATT'].rank(ascending = False)
```

```python
rbs['YDS/ATT_rank'] = rbs['YDS/ATT'].rank(ascending = False)
```

```python
rbs['ATT/G_rank'] = rbs['ATT/G'].rank(ascending = False)
```

```python
rbs['100+ YDS_rank'] = rbs['100+ YDS'].rank(ascending = False)
```

```python
rbs['TD_rank'] = rbs['TD'].rank(ascending = False)
```

```python
rbs['LONG_rank'] = rbs['LONG'].rank(ascending = False)
```

```python
rbs['FUM_rank'] = rbs['FUM'].rank()
```

```python
rbs['FUML_rank'] = rbs['FUML'].rank()
```

```python
rbs.to_csv('../Capstone_csv_file/rbs_stats_rank_19-20')
```

```python
rbs_rank = rbs.drop(['ATT', 'YDS/ATT', 'ATT/G', '100+ YDS', 'TD', 'LONG', 'FUM', 'FUML', 'YDS'], axis = 1)
```

```python
rbs_rank['overall'] = (rbs['ATT_rank'] + rbs['YDS/ATT_rank'] + rbs['ATT/G_rank'] +
                       rbs['100+ YDS_rank'] + rbs['TD_rank'] + rbs['LONG_rank'] + rbs['FUM_rank'] + 
                       rbs['FUML_rank'] + rbs['YDS_rank'])/9
```

```python
rbs_rank
```

```python
rbs_final_rank = {}
```

```python
rbs_final_rank['final_rank'] = rbs_rank['overall'].rank() 
```

```python
rbs_final_rank = pd.DataFrame(rbs_final_rank)
```

```python
rbs_final_rank = rbs_final_rank.sort_values('final_rank')
```

```python
rbs_final_rank.to_csv('../Capstone_csv_file/rbs_final_rank_19-20')
```
