# Tightends EDA

```python
import pandas as pd
```

```python
tes = pd.read_csv('../Capstone_Scrapping/Capstone_text_files/tight_end_17-18.txt')
```

```python
tes
```

```python
tes['YDS'] = tes['YDS'].str.replace(',', '').astype('int64')
```

```python
tes.rename(columns=lambda x: x.strip(), inplace = True)
```

```python
tes = tes[tes['TGT'] > 16].copy()
```

```python
tes
```

```python
tes = tes.groupby('NAME', sort = False).mean()
```

```python
tes.drop(columns = ['RTG', 'GP'], axis = 1, inplace = True)
```

```python
tes_stats = tes.copy()
```

```python
tes_rank = {}
```

```python
tes_rank['YDS_rank'] = tes['YDS'].rank(ascending = False)
```

```python
tes_rank['TGT_rank'] = tes['TGT'].rank(ascending = False)
```

```python
tes_rank['REC_rank'] = tes['REC'].rank(ascending = False)
```

```python
tes_rank['TGT/G_rank'] = tes['TGT/G'].rank(ascending = False)
```

```python
tes_rank['100+ YDS_rank'] = tes['100+ YDS'].rank(ascending = False)
```

```python
tes_rank['YDS/REC_rank'] = tes['YDS/REC'].rank(ascending = False)
```

```python
tes_rank['TD_rank'] = tes['TD'].rank(ascending = False)
```

```python
tes_rank['LONG_rank'] = tes['LONG'].rank(ascending = False)
```

```python
tes_rank['FUM_rank'] = tes['FUM'].rank()
```

```python
tes_rank['FUML_rank'] = tes['FUML'].rank()
```

```python
tes_rank['overall'] = (tes_rank['TGT_rank'] + tes_rank['REC_rank'] + tes_rank['TGT/G_rank'] + tes_rank['100+ YDS_rank'] +
            tes_rank['YDS/REC_rank'] + tes_rank['TD_rank'] + tes_rank['LONG_rank'] +
            tes_rank['FUM_rank'] + tes_rank['FUML_rank'] + tes_rank['YDS_rank'])/10
```

```python
tes_rank = pd.DataFrame(tes_rank)
```

```python
tes_final_rank = {}
```

```python
tes_final_rank['final_rank'] = tes_rank['overall'].rank()
```

```python
tes_final_rank = pd.DataFrame(tes_final_rank)
```

```python
tes_final_rank = tes_final_rank.sort_values('final_rank')
```

```python
tes_final_rank.to_csv('../Capstone_csv_file/tes_final_rank_17-18')
```

```python
tes_final_rank
```

```python
tes = tes_stats.merge(tes_rank, on = 'NAME')
```

```python
tes.to_csv('../Capstone_csv_file/tes_stats_rank_17-18')
```
